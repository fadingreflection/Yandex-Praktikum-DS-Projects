### Оценка эффективности обогащения руды

Исходя из ТЗ, формула для расчета эффективности обогащения рассчитывается по формуле:
recovery = (C x (F-T)) / (F x (C-T)) * 100, где:
- C - доля золота в концентрате после флотации/очистки (rougher.output.concentrate_au)
- F - доля золота в сырье/концентрате до флотации/очистки (rougher.input.feed_ag)
- T - доля золота в хвостах после флотации/очистки (rougher.output.tail_au)

```python
#сохраним данные в переменные
c, f, t = [data_train['rougher.output.concentrate_au'], 
data_train['rougher.input.feed_au'],
data_train['rougher.output.tail_au']]

#рассчитаем эффективность
recovery = (c * (f-t)) / (f * (c-t)) * 100

#рассчет МАЕ между recovery и дефолтными значениями
mae = mean_absolute_error(recovery, data_train['rougher.output.recovery'])
print('Средняя абсолютная ошибка между нашими и дефолтными расчетами равна:', mae)
```

```python
#визуализируем схожесть обогащения руды
sns.set(rc = {'figure.figsize':(7,4)})
plt.axvline(recovery.mean(), linestyle='--', color='green');
plt.axvline(recovery.median(), linestyle='--', color='red');
plt.title('Эффективность богащения руды согласно нашим расчетам')
sns.distplot(recovery)
plt.show()

sns.set(rc = {'figure.figsize':(7,4)})
plt.axvline(data_train['rougher.output.recovery'].mean(), linestyle='--', color='green');
plt.axvline(data_train['rougher.output.recovery'].median(), linestyle='--', color='red');
plt.title('Эффективность обогащения руды согласно дефолтным расчетам')
sns.distplot(data_train['rougher.output.recovery'])
plt.show()
```

<b>Промежуточный вывод:</b> наши расчеты не сильно отличаются от дефолтных. Значит, можно переходить к предобработке данных.

### Исследование суммарной концентрации всех веществ на разных стадиях

```python
stages = ['rougher.output.concentrate_', 'primary_cleaner.output.concentrate_',
         'final.output.concentrate_']

for stage in stages:
    stage_data = data_full.filter(regex=stage)
    stage_data['substance_sum'] = stage_data.sum(axis = 1)
    sns.distplot(stage_data['substance_sum'], label = stage)
    plt.title('Суммарная концентрация всех веществ на выходе разных стадий очистки')
    plt.legend()
```

<b>Промежуточный вывод:</b> концентрация веществ стабильно растет на каждом этапе очистки, однако видно значительное кол-во нулей, чего быть не может, поскольку что-то из веществ должно быть в суммарном содержании. Избавимся от этих аномалий и удалим нули.

```python
#посмотрим на суммарную концентрацию веществ на стадиях
#с учетом убранных нулей
stages = ['rougher.output.concentrate_', 'primary_cleaner.output.concentrate_',
         'final.output.concentrate_']

for stage in stages:
    stage_data = data_full.filter(regex=stage)
    stage_data['substance_sum'] = stage_data.sum(axis = 1)
    #избавляемся от нулевых значений
    stage_data['substance_sum'] = stage_data['substance_sum'].replace(0, np.nan)
    stage_data['substance_sum'].dropna(inplace = True)
    #строим графики
    sns.distplot(stage_data['substance_sum'], label = stage)
    plt.title('Суммарная концентрация всех веществ на выходе разных стадий очистки без нулей')
    plt.legend()
```

Готово. Теперь напишем ф-цию расчета sMAPE.

### Создание функции sMAPE
Согласно ТЗ, формула выглядит как:
100 / len(ЦП) * sum(2 * abs(предсказание - ЦП) / (abs(ЦП) + abs(предсказание)))

```python
#создадим ф-цию расчета sMAPE
def sMAPE(y, predictions):
    return 1 / len(y) * sum(2 * abs(predictions - y) / (abs(y) + abs(predictions))) * 100

#и для расчета final sMAPE
def sMAPE_final(sMAPE_rougher, sMAPE_final):
    return 0.25 * sMAPE_rougher + 0.75 * sMAPE_final
```
